```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def func(x: float) -> float:
    return x**2 + 5
```

```python
x1 = np.linspace(-1, 1, 200)
x2 = np.linspace(-6, 6, 200)
x3 = np.linspace(0, 5, 200)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 5))
ax1.plot(x1, func(x1), label="function")
ax1.set_title("(-1, 1)")
ax1.legend()
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.grid()

ax2.plot(x2, func(x2), label="function")
ax2.set_title("(-6, 6)")
ax2.legend()
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.grid()

ax3.plot(x3, func(x3), label="function")
ax3.set_title("(0, 5)")
ax3.legend()
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.grid()
plt.show()
```
